```python
#Explore the dataset
#Disaster Categories
import pandas as pd
categories_df = pd.read_csv('disaster_categories.csv')
```

```python
categories_df.head()
```

```python
#Looks like there are duplicate sets of ids as the  count of id more than the count of unique ids 
categories_df.shape, categories_df.id.nunique(), categories_df.id.isna().sum()
```

```python
# Find duplicate sets of rows with matching ids and categories text
len(categories_df[categories_df.duplicated(keep='first')])
```

```python
# Deduplicate the dataframe by deleting the duplicate rows
duplicate_index = categories_df[categories_df.duplicated(keep='first')].index
duplicate_index
# drop these row indexes from dataFrame 
categories_df.drop(duplicate_index, axis=0, inplace = True) 
```

```python
categories_df.shape, categories_df.id.nunique(), categories_df.id.isna().sum()
```

```python
# Check for duplicated ids with non-duplicated categories
len(categories_df[categories_df.id.duplicated(keep='first')]),categories_df[categories_df.id.duplicated(keep='first')]
```

```python
categories_df[categories_df.id==202]
```

```python
# Conver the categories to individual columns and then find duplicates
# Using a simple for loop since number of records are not many
for row in categories_df.iloc:
    cats = [cat.strip().split('-') for cat in row.categories.split(';')]
    for cat, value in cats:
        categories_df[cat]=int(value)
```

```python
categories_df.shape
```

```python
categories_df.drop(columns=['categories'], inplace=True)
```

```python
categories_df.id.nunique(), len(categories_df)
```

```python
categories_df[categories_df.duplicated(keep='first')]
```

```python
# We can now drop the duplicate rows again
# Deduplicate the dataframe by deleting the duplicate rows
duplicate_index = categories_df[categories_df.duplicated(keep='first')].index
duplicate_index
# drop these row indexes from dataFrame 
categories_df.drop(duplicate_index, axis=0, inplace = True) 
```

```python
categories_df.id.nunique(), len(categories_df)
```

```python
# Check the disaster messages csv
messages_df = pd.read_csv('disaster_messages.csv')
```

```python
messages_df.head()
```

```python
#Looks like there are duplicate sets of ids as the  count of id more than the count of unique ids 
messages_df.shape, messages_df.id.nunique(), messages_df.id.isna().sum()
```

```python
messages_df[messages_df.duplicated()]
```

```python
#Remove duplicate rows in messages_df
duplicate_index = messages_df[messages_df.duplicated(keep='first')].index
duplicate_index
# drop these row indexes from dataFrame 
messages_df.drop(duplicate_index, axis=0, inplace = True) 
```

```python
messages_df.shape, messages_df.id.nunique()
```

```python
#We can now merge the two dataframes
consolidated_df = pd.merge(left=categories_df, right=messages_df, how='left', left_on='id', right_on='id')
```
